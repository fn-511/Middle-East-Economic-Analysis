1. Data Loading and Initialization

In this section, we import the necessary libraries (Pandas, Matplotlib, and Seaborn) and load the Middle East economic dataset covering the period from 1990 to 2024.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Middle_East_Economic_Data_1990_2024_with_Oil.csv')
df.head()
```

2. Data Cleaning and Inspection

We inspect the data structure and handle missing values using dropna() to ensure the accuracy of our statistical analysis and visualizations.

```python
df.info()
```

```python
df =df.dropna()
```

3. Case Study: Saudi Arabia's GDP Trend

Filtering the data for Saudi Arabia to visualize its GDP growth over the decades, providing a baseline for the region's largest economy.


```python
sa = df[df['Country']=='Saudi Arabia']
sa
```

```python
sns.lineplot(x = 'Year',y='GDP_current_USD' ,data = df )
plt.title('Saudi Araabia')
plt.show()
```

4. Regional Comparison: Top 5 Economies
Comparing the average GDP across Middle Eastern countries to identify the top 5 economic powers in the region.

```python
top5 = df.groupby('Country')['GDP_current_USD'].mean().sort_values(ascending=False).head().reset_index()
```

```python
sns.barplot(x = 'GDP_current_USD',y = 'Country',data = top5 )

plt.title('Top 5 Middle Eastern Countries by Average GDP')
plt.xlabel('Average GDP (Current USD)')
plt.ylabel('Country')
```

6. Growth vs. Volatility

Examining the relationship between average GDP growth and economic volatility. This helps identify which countries achieve stable growth versus those with high fluctuations.


```python
mean_growth = df.groupby('Country')['GDP_growth_annual_pct'].mean()

std_growth = df.groupby('Country')['GDP_growth_annual_pct'].std()
growth_df = pd.DataFrame({
    'Mean_Growth': mean_growth,
    'Volatility': std_growth
}).reset_index()
sns.scatterplot(
    x='Mean_Growth',
    y='Volatility',
    hue='Country',
    data=growth_df
)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Economic Growth vs Volatility')
plt.xlabel('Average GDP Growth (%)')
plt.ylabel('Growth Volatility (Standard Deviation)')
plt.show()
```

7. Export-Driven Economies

Investigating the share of exports in the GDP and its relationship with the overall economic size of each country.


```python
for i in df['Country'].unique():
    cn = df[df['Country']==i]
    ar=cn['Exports_pct_GDP'].corr(cn['GDP_current_USD'])
    
summary = df.groupby('Country')[['Exports_pct_GDP', 'GDP_current_USD']].mean().reset_index()
sns.scatterplot(x = 'Exports_pct_GDP',y='GDP_current_USD',data =summary,hue = 'Country' )
plt.legend(bbox_to_anchor=(1.05, 1))
plt.title('Exports Share vs. Total Economic Size')
plt.ylabel('Total GDP (Current USD)')
plt.xlabel('Exports (% of GDP)')
plt.show()
```

```python
cg =df.groupby('Country')['GDP_growth_annual_pct'].mean().sort_values().reset_index()
sns.barplot(x = 'GDP_growth_annual_pct',y='Country',data=cg)

plt.title('Average Annual GDP Growth Rate by Country')
plt.xlabel('GDP Growth (%)')
plt.show()
```

8. Foreign Investment and the Labor Market

Analyzing whether Foreign Direct Investment (FDI) inflows have a measurable impact on reducing unemployment rates across the region.


```python
fu = df.groupby('Country')[['Unemployment_total_pct','FDI_net_inflows_pct_GDP']].mean().reset_index()
sns.scatterplot(x = 'FDI_net_inflows_pct_GDP',y ='Unemployment_total_pct',hue= 'Country',data = fu)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Relationship between FDI Inflows and Unemployment Rates')
plt.xlabel('Foreign Direct Investment, net inflows (% of GDP)')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

9. Economic Efficiency: Growth vs. Inflation

Mapping countries based on their ability to achieve high growth while maintaining manageable inflation levels.


```python
gi = df.groupby('Country')[['GDP_growth_annual_pct','Inflation_consumer_prices_annual_pct']].mean().reset_index()
sns.scatterplot(x = 'GDP_growth_annual_pct',y = 'Inflation_consumer_prices_annual_pct',data = gi,hue = 'Country')
plt.xlabel('Average Annual GDP Growth (%)')
plt.ylabel('Average Inflation Rate (%)')
plt.title('Economic Efficiency: Annual GDP Growth vs Inflation Rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axvline()
plt.axhline()
plt.show()
```

10. Detailed Country Resilience Trends
A comprehensive look at the annual GDP growth rates for every country individually to identify specific periods of crisis and recovery (e.g., 2008 and 2020).



```python
n = len(df.Country.unique())
plt.figure(figsize=(20,30))
for i,country in enumerate(df.Country.unique(),1):
    plt.subplot(n,3,i)
    country_data=(df[df['Country']==country])
    sns.lineplot(data = country_data,x = 'Year',y = 'GDP_growth_annual_pct',marker = 'o')
    plt.title(f'Economic Growth Resilience : {country}')
    plt.grid(True)



plt.tight_layout()
plt.show()
```
